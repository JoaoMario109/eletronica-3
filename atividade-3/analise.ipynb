## Análise Teórica
Nesta seção, serão apresentados os resultados da análise teórica do oscilador astável projetado. 

Este circuito pode ser dividido em dois blocos, o primeiro que é um integrador que tera como entrada uma constante o que resultara em uma rampa teoricamente infinita na saida e o bloco comparador que irá comparar a tensão da rampa com um valor DC e quando a rampa superar o valor DC irá limpar o integrador, voltando o mesmo ao valor inicial.

### Circuito integrador 

Este circuito ira integrar a entrada (constante) para gerar uma rampa na saida. A seguir sera apresentado a dedução de como funciona o mesmo

<img width=50% src="./assets/integrator.png"/>

*Circuito Integrador*

Usando lei dos nós:
$$I_{c} + I_{r} + I_{op} = 0$$

A corrente no ampop é teoricamente 0
$$I_{op} = 0$$

A corrente em um resistor é
$$I_{r} = V_{r} / R1$$
$$V_{r} = V_{in} - V_{-}$$
$$I_{r} = \frac{V_{in} - V_{-}}{R1}$$

A corrente em um capacitor é
$$
I_{c} = C\frac{dv}{dt} \\
V_{c} = V_{out} - V_{-} \\
I_{c} = C\frac{d(V_{out} - V_{-})}{dt}
$$

Usando a propiedade do *curto virtual*
$$V_{-} = V_{+}$$
$$V_{+} = 0$$
logo
$$V_{-} = 0$$

Substituindo na lei dos nós

$$
C\frac{d(V_{out})}{dt} + \frac{V_{in}}{R1} = 0
$$

Integrando

$$
\int{\frac{V_{in}}{R1}} + V_{out} * C = 0
$$

Isolando
$$
\int{\frac{V_{in}}{R1}} = -V_{out} * C
$$

Como V_{in} é um valor DC (constante)
$$
\int{\frac{V_{in}}{R1}} = \frac{V_{in} * t}{R1}
$$

Isolando V_{out}
$$
V_{out} = \frac{-V_{in} * t}{R1 * C}
$$





### Circuito comparador

<img width=50% src="./assets/comparador.png"/>

O circuito é basicamente o integrador, somente que o $V_{+}$ varia


*Circuito Integrador*

Usando lei dos nós:
$$I_{c} + I_{r} + I_{op} = 0$$

A corrente no ampop é teoricamente 0
$$I_{op} = 0$$

A corrente no resistor é
$$I_{r} = \frac{V_{in} - V_{-}}{R1}$$

A corrente no capacitor é
$$
I_{c} = C\frac{d(V_{out} - V_{-})}{dt}
$$

Usando a propiedade do *curto virtual*
$$V_{-} = V_{+}$$
$$V_{+} = V_{dc}$$
logo
$$V_{-} = V_{dc}$$

Substituindo na lei dos nós

$$
C\frac{d(V_{out - V_{dc}})}{dt} + \frac{V_{in - V_{dc}}}{R1} = 0
$$

Integrando

$$
\int{\frac{V_{in} - V_{dc}}{R1}} + (V_{out} - V_{dc}) * C = 0
$$

Isolando
$$
\int{\frac{V_{in} - V_{dc}}{R1}} = -(V_{out} - V_{dc}) * C
$$

Como V_{in} é um valor DC (constante) e Vdc também
$$
\int{\frac{V_{in} - V_{dc}}{R1}} = \frac{(V_{in} - V_{dc}) * t}{R1}
$$

Isolando V_{out}
$$
V_{out} = \frac{-(V_{in} - V_{dc}) * t}{R1 * C} + V_{dc}
$$

### Circuito total

O primeiro circuito basicamente gera uma rampa de 0 até Vdc já que o circuito comparador, descarrega o capacitor integrador quando $V_{out} = V_{dc}$ então a frequencia dessa rampa é
$$
V_{out} = \frac{-V_{in} * t}{R1 * C}
$$

$$
freq = \frac{-V_{in}}{V_{dc} * R1 * C}
$$

Como o valor de $ R1 * C$ é muito baixo no circuito comparador, a constante que multiplica t é muito alta em relação a frequencia de oscilação, então o comportamento desse circuito é um comparador que tem um pequeno delay ao ligar e desligar, conseguindo manter um tempo suficiente para que o transistor consiga descarregar totalmente o capacitor do integrador

Calculando para os valores do circuito:

$$
freq = \frac{-V_{in}}{V_{dc} * 126.9e-6}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import json
```

```python
measurements_name = ["vmax", "vmin", "vpp", "vavg", "frequency", "pslewrate", "nslewrate", "vmid",]
channels = ["CHAN1", "CHAN2", "CHAN3", "CHAN4"]
channels_name = {"CHAN1": "Vtransitor_base", "CHAN2": "Vout", "CHAN3": "Vco", "CHAN4": "Vdc"}

def get_waveforms(test_name):
    return pd.read_csv(test_name + '/waveforms.csv', index_col='time')

def get_all_measurements(test_name):
    with open(test_name + '/measurements.json') as json_file:
        return json.load(json_file)
    
def get_theoretical_frequency(vco, vdc):
    return -vco / (vdc * 100e3 * 2.7e-9)
    
```

#### Teste da variação de vco

Para o teste foi mantido Vdc fixo em 6V e variado Vco para os valores -2V -6V e -10V

```python
# Comparing waveforms 
df1 = get_waveforms("vco--2,vdc-6")
df2 = get_waveforms("vco--6,vdc-6")
df3 = get_waveforms("vco--10,vdc-6")

# Plot waveforms using subplots
fig, axs = plt.subplots(4, 1, figsize=(10, 10))
for i, channel in enumerate(channels):  
	axs[i].plot(df1.index, df1[channel], label='vco: -2V')
	axs[i].plot(df2.index, df2[channel], label='vco: -4V')
	axs[i].plot(df3.index, df3[channel], label='vco: -10V')
	axs[i].set_title(channels_name[channel])
	axs[i].set_xlabel('Time (s)')
	axs[i].set_ylabel('Voltage (V)')
	axs[i].legend()
	axs[i].grid(True)

	# Increase space between subplots
	fig.subplots_adjust(hspace=0.5)
	
	# Limit x axis to values between -0.001 and 0.001
	axs[i].set_xlim([-0.001, 0.001])

	# Configure scale to plot between min value and max voltage
	# If min value is greater than 0 plot between 0 and max voltage
	if df1[channel].min() > 0:
		axs[i].set_ylim([0, df1[channel].max() + 0.1])

vco_2 = get_all_measurements("vco--2,vdc-6")
vco_6 = get_all_measurements("vco--6,vdc-6")
vco_10 = get_all_measurements("vco--10,vdc-6")

print("Vco = -2v, Vdc = 6v")
print("teorico: ", get_theoretical_frequency(-2, 6))
print("pratico: ", vco_2["CHAN2"]["frequency"])
print("Vco = -6v, Vdc = 6v")
print("teorico: ", get_theoretical_frequency(-6, 6))
print("pratico: ", vco_6["CHAN2"]["frequency"])
print("Vco = -10v, Vdc = 6v")
print("teorico: ", get_theoretical_frequency(-10, 6))
print("pratico: ", vco_10["CHAN2"]["frequency"])
```
